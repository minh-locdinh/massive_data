```
from pyspark import SparkConf, SparkContext
import collections

conf = SparkConf().setMaster("local").setAppName("WordCount")
sc = SparkContext.getOrCreate(conf=conf)
```

```
downloaded = drive.CreateFile({'id':'1oE0UFWLmc966djXfYtCa9t4s8uiFfyRa'})
downloaded.GetContentFile('BaiTap1.txt')
```

```
file = open("BaiTap1.txt", "r")
text = file.read()
print(text)
```

**Count total words**

```
text_input = text.split()
```

```
rdd = sc.parallelize(text_input)
rdd.count()
```

**Print most common words (top 10)**

```
counts = rdd.map(lambda word:(word,1))
xyz = counts.reduceByKey(lambda x,y:(x+y))
print(xyz.collect())
```

```
from collections import Counter
word_counts = Counter(text_input)
most_common_words = word_counts.most_common(10)
print(most_common_words)
```
