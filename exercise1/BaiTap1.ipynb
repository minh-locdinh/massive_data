```
from pyspark import SparkConf, SparkContext
import collections

conf = SparkConf().setMaster("local").setAppName("WordCount")
sc = SparkContext.getOrCreate(conf=conf)
```

**Read & count total words**

```
text = """co trai tim moi biet yeu moi biet vui moi bik buon 
bik nho nhung co nhung dung than tho hat mot minh. 
ngay dau gap nhau oi sao long thay nao nao 
nhung van vuong den nhe nhang tim nhau nghe boi hoi that lau phut ban dau. nhung dau yeu ay thoang qua cu ngo nhu moi hom nao, 
co trai tim da bik mo da bik mong da bik cho. 
tim dau e den sao qua nhanh nhung nho thuong da qua nhanh den khong ngo 
gio e vui cuoi that hon nhien phut mong cho. 
luc moi yeu that vui bik bao nhiu ngay thang moi yeu dep oi bik bao nhiu, 
e so ngay buon lai mang tinh di mat nen moi khi minh vui phut ben nhau 
e van nam tay nhu luc moi quen nhau hom nao uoc cho minh mai nhu ban dau, dep thay tieng yeu dau"""
text_input = text.split()
rdd = sc.parallelize(text_input)
rdd.count()
```

**Count words & print most common words (10 words)**

```
counts = rdd.map(lambda word:(word,1))
xyz = counts.reduceByKey(lambda x,y:(x+y))
print(xyz.collect())
```

```
from collections import Counter
word_counts = Counter(text_input)
most_common_words = word_counts.most_common(10)
print(most_common_words)
```
